```python
#additional dictionaries
TNFA4_visit_dict = {}
for visit in lomas_TNFA.VisitLabel.unique():
    TNFA4_visit_dict[visit] = lomas_TNFA_4[
        lomas_TNFA_4.VisitLabel == visit
    ].ValueNum
    
TNFA5_visit_dict = {}
for visit in lomas_TNFA.VisitLabel.unique():
    TNFA5_visit_dict[visit] = lomas_TNFA_5[
        lomas_TNFA_5.VisitLabel == visit
    ].ValueNum
    
TNFA6_visit_dict = {}
for visit in lomas_TNFA.VisitLabel.unique():
    TNFA6_visit_dict[visit] = lomas_TNFA_6[
        lomas_TNFA_6.VisitLabel == visit
    ].ValueNum
    
for visit in lomas_TNFA.VisitLabel.unique():
    statistic, p_value = stats.f_oneway(TNFA4_visit_dict[visit], TNFA5_visit_dict[visit], TNFA6_visit_dict[visit])

    # Display the results
    print("ANOVA Statistic:", statistic)
    print("P-value:", p_value)

    # Check the significance level (e.g., 0.05)
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis. There are significant differences between groups.")
    else:
        print("Fail to reject the null hypothesis. There are no significant differences between groups.")
```
